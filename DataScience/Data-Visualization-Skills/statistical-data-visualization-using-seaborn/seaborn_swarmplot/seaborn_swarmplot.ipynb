```python
import seaborn as sns
import warnings
```

```python
warnings.filterwarnings('ignore')
sns.set_theme(style="whitegrid", palette="muted")
```

```python
penguins=sns.load_dataset('penguins').dropna()
penguins.head(3)
```

```python
ax = sns.swarmplot(
    data=penguins, 
    x="body_mass_g", 
    y="sex", 
    hue="species"
)
ax.set(ylabel="")
```

```python
import seaborn as sns
import warnings
import numpy as np
```

```python
sns.set_theme(style='ticks')
warnings.filterwarnings('ignore')
```

```python
rs = np.random.RandomState(100)
x = rs.gamma(2, size=1000)
y = -.5 * x + rs.normal(size=1000)
z = rs.normal(size=1000)
```

```python
sns.jointplot(
    x=x, 
    y=y, 
    kind="hex", 
    color="#4CB391"
)
```

```python
sns.jointplot(
    x=x, 
    y=z, 
    kind="hex", 
    color="#4CB391"
)
```

```python
penguins=sns.load_dataset('penguins').dropna()
penguins.head(3)
```

```python
sns.jointplot(
    data=penguins,
    x='bill_length_mm',
    y='body_mass_g',
    hue='species',
)
```

```python
sns.jointplot(
    data=penguins,
    x='bill_length_mm',
    y='body_mass_g',
    hue='species',
    kind='kde'
)
```

```python
tips=sns.load_dataset('tips').dropna()
tips.head(3)
```

```python
sns.jointplot(
    data=tips,
    x='total_bill',
    y='tip',
    #hue='sex',
    xlim=(0, 60),
    ylim=(0,12),
    height=7,
    color='m',
    kind='reg',
    truncate=False
)
```

```python
import seaborn as sns
import warnings
import numpy as np
```

```python
warnings.filterwarnings('ignore')
sns.set_theme(style='whitegrid')
```

```python
rs = np.random.RandomState(1)
x = rs.normal(2, 1, 75)
y = 2 + 1.5 * x + rs.normal(0, 2, 75)
```

```python
sns.residplot(
    x=x,
    y=y,
    color="g"
)
```

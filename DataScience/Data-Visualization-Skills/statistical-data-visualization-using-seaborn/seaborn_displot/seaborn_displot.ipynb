```python
import seaborn as sns
import warnings
```

```python
warnings.filterwarnings('ignore')
sns.set_theme(style = 'darkgrid')
```

```python
penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
penguins.shape
```

```python
penguins=penguins.dropna()
```

```python
penguins.shape
```

```python
penguins.nunique()
```

```python
sns.displot(
    data=penguins,
    x='flipper_length_mm',
    col='species',
    row='sex',
    binwidth=3,
    height=5,
    kde=True,
    facet_kws=dict(margin_titles=True)
)
```

```python
sns.displot(
    data=penguins,
    x='body_mass_g',
    hue='species',
    binwidth=50
)
```

```python
sns.displot(
    data=penguins,
    x='body_mass_g',
    binwidth=50,
    col='species'
)
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.displot(
    data=tips,
    x='total_bill',
    y='tip',
    col='sex',
    row='smoker',
    hue='day',
    kind='kde',
    facet_kws=dict(margin_titles=True)
)
```

```python
sns.displot(
    data=tips,
    x='tip',
    kind='hist'
)
```

```python
sns.displot(
    data=tips,
    x='tip',
    kind='kde'
)
```

```python
sns.displot(
    data=tips,
    x='tip',
    kde=True
)
```

```python
diamonds=sns.load_dataset('diamonds')
diamonds.head()
```

```python
sns.displot(
    data=diamonds,
    x='price',
    y='color',
    col='clarity',
    log_scale=(True, False),
    col_wrap=4,
    height=4,
    aspect=.7
)
```

```python
sns.set_theme(style='whitegrid')
```

```python
sns.displot(
    data=diamonds,
    x='carat',
    hue='cut',
    height=6,
    kind='kde',
    multiple='fill',
    clip=(0,None),
    palette="ch:rot=-.25,hue=1,light=.75"
)
```

```python
sns.set_theme(style='ticks')
```

```python
cars=sns.load_dataset('mpg')
cars.head(3)
```

```python
colors = (250, 70, 50), (350, 70, 50)
cmap = sns.blend_palette(colors, input="husl", as_cmap=True)
sns.displot(
    cars,
    x="displacement", col="origin", hue="model_year",
    kind="ecdf", aspect=.75, linewidth=2, palette=cmap,
)
```

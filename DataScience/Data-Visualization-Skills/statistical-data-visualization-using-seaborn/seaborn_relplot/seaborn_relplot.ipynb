```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

seaborn에서 제공하는 default 데이터 파일 mpg를 로드한다. 

```python
cars = sns.load_dataset("mpg")
```

데이터의 자료형이 데이터 프레임인것을 확인한다.

```python
#type(cars)
#cars.info()
cars.nunique()
```

데이터의 정보를 보고 컬럼 별 자료형과 nan이 없음을 확인한다. 

만약에 대비해서 nan 자료들을 데이터에서 삭제한다.

```python
cars.dropna(inplace = True)
cars.head()
```

불필요한 deprecated warnings 들이 안 뜨도록 warnings 라이브러리를 사용하고 원하는 그래픽을 출력한다. 

```python
warnings.filterwarnings('ignore')
sns.relplot(x='model_year', y='mpg', col='origin', hue='cylinders', data=cars)
```

```python
bn=sns.load_dataset('brain_networks', header=[0,1,2], index_col=0)
```

```python
used_networks = [1, 5, 6, 7, 8, 12, 17]
used_columns = (bn.columns
                  .get_level_values("network")
                  .astype(int)
                  .isin(used_networks))
bn = bn.loc[:, used_columns]
bn.columns = bn.columns.map("-".join)

corr_mat = bn.corr().stack().reset_index(name="correlation")
```

```python
g = sns.relplot(
    data=corr_mat,
    x="level_0", y="level_1", hue="correlation", size="correlation",
    palette="vlag", hue_norm=(-1, 1), edgecolor=".7",
    height=10, sizes=(50, 250), size_norm=(-.2, .8),
)
g.set(xlabel="", ylabel="", aspect="equal")
g.despine(left=True, bottom=True)
g.ax.margins(.02)
for label in g.ax.get_xticklabels():
    label.set_rotation(90)
```

```python
planets = sns.load_dataset("planets")
planets.head()
```

```python
cmap = sns.cubehelix_palette(rot=-.2, as_cmap=True)
```

```python
g = sns.relplot(
    data=planets,
    x="distance", y="orbital_period",
    hue="year", size="mass",
    palette=cmap, sizes=(10, 200),
)
g.set(xscale="log", yscale="log")
g.ax.xaxis.grid(True, "minor", linewidth=.25)
g.ax.yaxis.grid(True, "minor", linewidth=.25)
g.despine(left=True, bottom=True)
```

```python
sns.set_theme(style="dark")
```

```python
flights = sns.load_dataset("flights")
flights.head()
```

```python
g = sns.relplot(
    data=flights,
    x="month", 
    y="passengers", 
    col="year", 
    hue="year",
    kind="line", 
    palette="crest", 
    linewidth=4, 
    zorder=5,
    col_wrap=4, 
    height=2, 
    aspect=1.5, 
    legend=False,
)
for year, ax in g.axes_dict.items():
    ax.text(.8, .85, year, transform=ax.transAxes, fontweight="bold")
    sns.lineplot(
        data=flights, x="month", y="passengers", units="year",
        estimator=None, color=".7", linewidth=1, ax=ax,
    )
ax.set_xticks(ax.get_xticks()[::2])

g.set_titles("")
g.set_axis_labels("", "Passengers")
g.tight_layout()
```

```python
import seaborn as sns
import warnings
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
warnings.filterwarnings('ignore')
```

```python
penguins = sns.load_dataset('penguins').dropna()
penguins.head()
```

```python
sns.set_style('darkgrid')
```

```python
sns.histplot(
    data=penguins,
    x='bill_length_mm'
)
```

```python
sns.histplot(
    data=penguins,
    x='bill_length_mm',
    kde=True
)
```

```python
sns.histplot(
    data=penguins,
    x='bill_length_mm',
    binwidth=1,
    binrange=(30,60)
)
```

```python
#sns.histplot(x='bill_length_mm', data=penguins, stat='density')
#sns.histplot(x='bill_length_mm', data=penguins, stat='probability')
#sns.histplot(x='bill_length_mm', data=penguins, stat='probability', fill=False)
#sns.histplot(x='bill_length_mm', data=penguins, stat='probability', fill=False, cumulative=True)
sns.histplot(x='bill_length_mm', data=penguins, stat='probability', fill=False, cumulative=True, element='step')
```

```python
#sns.histplot(x='bill_length_mm', data = penguins, hue='species')
#sns.histplot(x='bill_length_mm', data = penguins, hue='species', element='step')
#sns.histplot(x='bill_length_mm', data = penguins, hue='species', element='poly')
#sns.histplot(x='bill_length_mm', data = penguins, hue='species', multiple='stack')
sns.histplot(x='bill_length_mm', data = penguins, hue='species', multiple='fill')
```

```python
sns.histplot(
    data=penguins,
    x='bill_length_mm',
    y='bill_depth_mm',
    cbar=True
)
```

```python
sns.histplot(
    data=penguins,
    x='bill_length_mm',
    y='bill_depth_mm',
    hue='species',
)
```

## styling

```python
sns.histplot(
    data=penguins,
    x='species',
    hue='sex',
    shrink=.8,
    palette='bone',
    multiple='dodge'
)
```

```python
sns.histplot(
    data=penguins,
    x='bill_length_mm',
    fill=False
)
```

```python
sns.set_theme(style='ticks')
```

```python
diamonds=sns.load_dataset('diamonds').dropna()
diamonds.head()
```

```python
f,ax=plt.subplots(figsize=(7,6))
sns.despine(f)
sns.histplot(
    data=diamonds,
    x='price',
    hue='cut',
    multiple='stack',
    log_scale=True,
)
ax.xaxis.set_major_formatter(mpl.ticker.ScalarFormatter())
ax.set_xticks([500, 1000, 2000, 5000, 10000])
```

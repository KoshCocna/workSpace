```python
import seaborn as sns
import warnings
```

```python
warnings.filterwarnings('ignore')
sns.set_theme(style='whitegrid')
```

```python
penguins=sns.load_dataset('penguins').dropna()
#penguins.shape
penguins.nunique()
```

```python
penguins.head()
```

```python
g=sns.catplot(
    data=penguins,
    x='species',
    y='body_mass_g',
    hue='sex',
    palette='dark',
    height=4,
    alpha=.6,
    errorbar='sd'
)
g.despine(left=True)
g.set_axis_labels('','Body Mass(g)')
g.legend.set_title('catplot-example')
```

```python
cars = sns.load_dataset("mpg")
cars.head()
```

```python
sns.catplot(
    data=cars, 
    x="cylinders", 
    y="acceleration", 
    hue="weight",
    native_scale=True, 
    zorder=1
)
sns.regplot(
    data=cars, 
    x="cylinders", 
    y="acceleration",
    scatter=False, 
    truncate=False, 
    order=2, 
    color=".2",
)
```

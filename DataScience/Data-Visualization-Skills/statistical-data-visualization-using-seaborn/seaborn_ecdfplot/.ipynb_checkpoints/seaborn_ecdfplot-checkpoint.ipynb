```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
tips = sns.load_dataset("tips").dropna()
tips.head()
```

```python
warnings.filterwarnings('ignore')
plt.subplot(2,2,1)
sns.ecdfplot(x='tip', data=tips)
plt.axvline(x=4,c='blue')
plt.subplot(2,2,2)
sns.ecdfplot(x='tip', data=tips, weights='tip')
plt.axhline(y=0.5,c='red')
```

The graphic above shows that 80% of people tip 4 or less.  
Use weights to weigh the proportion, here 50% of tip money made came from tips less than about 3.5

### Additional Options

```python
graph = sns.ecdfplot(x='tip', data=tips, hue='sex', palette='summer', lw=2)
graph.legend(['male','female'], title='Gender')
```

```python
p = sns.ecdfplot(x='tip', data=tips, hue='day', lw=1)
p.legend(['Thurday','Friday','Saturday','Sunday'], title='Day of Week', loc=4)
```

```python
sns.ecdfplot(x='tip', data=tips, stat='count', hue='day')
```

```python
import seaborn as sns
import warnings
import pandas as pd
```

```python
warnings.filterwarnings('ignore')
sns.set_theme(style='whitegrid')
```

```python
diamonds=sns.load_dataset('diamonds').dropna()
diamonds.head(3)
diamonds.info()
```

```python
sns.boxenplot(
    data=diamonds,
    x='clarity',
    y='carat',
    color='g'
)
```

```python
sns.boxenplot(
    data=diamonds,
    x='price',
    y='clarity'
)
```

```python
diamonds.carat.unique()
```

```python
mega_diamonds=diamonds['carat'].gt(1).rename('mega_diamonds')
type(mega_diamonds)
```

```python
diamonds=pd.concat([diamonds, mega_diamonds], axis=1)
diamonds
```

```python
sns.boxenplot(
    data=diamonds,
    x='price',
    y='clarity',
    hue='mega_diamonds'
)
```

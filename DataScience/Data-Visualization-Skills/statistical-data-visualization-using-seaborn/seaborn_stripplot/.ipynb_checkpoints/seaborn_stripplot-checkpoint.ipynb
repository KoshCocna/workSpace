```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
sns.set_theme(style='whitegrid')
warnings.filterwarnings('ignore')
```

```python
iris=sns.load_dataset('iris').dropna()
iris.head()
```

```python
iris.nunique()
```

```python
sns.stripplot(
    data=iris,
    x='sepal_length',
    y='species',
    size=4,
    color='.1'
)
```

```python
iris=iris.melt(
    id_vars='species',
    var_name='measurement'
)
iris.head()
iris.nunique()
```

```python
fig, ax = plt.subplots()
sns.despine(left=True, bottom=True)

sns.stripplot(
    data=iris,
    x='value',
    y='measurement',
    hue='species',
    size=4,
    zorder=1,
    alpha=.5,
    dodge=True,
    legend=False
)
sns.pointplot(
    data=iris,
    x='value',
    y='measurement',
    hue='species',
    dodge=True,
    palette='dark',
    errorbar=None,
    markers='d',
    linestyles="none"
)

sns.move_legend(
    ax, loc="lower right", ncol=3, frameon=True, columnspacing=1, handletextpad=0,
)
```

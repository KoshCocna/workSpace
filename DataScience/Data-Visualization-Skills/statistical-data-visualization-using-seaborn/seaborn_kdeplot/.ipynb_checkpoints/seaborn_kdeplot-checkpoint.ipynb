## Seaborn: KDE

```python
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
cars = sns.load_dataset("mpg").dropna()
cars.shape
```

```python
cars.head()
```

## Univariate (1D)

```python
sns.kdeplot(cars.horsepower)
```

```python
sns.kdeplot(cars.horsepower, shade=True)
```

```python
sns.kdeplot(data=cars.horsepower, bw_adjust=0.3)
```

```python
sns.kdeplot(cars.horsepower, cumulative=True)
```

## Bivariate 2D

```python
sns.kdeplot(x=cars.horsepower, y=cars.mpg)
```

```python
sns.kdeplot(x=cars.horsepower, y=cars.mpg, n_levels=20)
```

```python
sns.kdeplot(x=cars.horsepower, y=cars.mpg, shade=True)
```

```python
sns.kdeplot(x=cars.horsepower, y=cars.mpg, shade=False, hue=cars.cylinders)
```

```python
sns.set_theme(style='white')
rs=np.random.RandomState(50)
rs
```

```python
f, axes = plt.subplots(3, 3, figsize=(9, 9), sharex=True, sharey=True)
for ax, s in zip(axes.flat, np.linspace(0, 3, 10)):
    cmap = sns.cubehelix_palette(start=s, light=1, as_cmap=True)
    x, y = rs.normal(size=(2, 50))
    sns.kdeplot(
        x=x, y=y,
        cmap=cmap, fill=True,
        clip=(-5, 5), cut=10,
        thresh=0, levels=15,
        ax=ax,
    )
    ax.set_axis_off()
ax.set(xlim=(-3.5, 3.5), ylim=(-3.5, 3.5))
f.subplots_adjust(0, 0, 1, 1, .08, .08)
```

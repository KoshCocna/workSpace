```python
from matplotlib import pyplot as plt
import seaborn as sns
import warnings
```

## 1th example with 'mpg' dataset

```python
warnings.filterwarnings('ignore')
cars=sns.load_dataset('mpg').dropna()
cars.head(6)
```

```python
cars.nunique()
```

```python
sns.boxplot(data=cars, x='model_year', y='mpg', hue='origin')
```

## 2th example with 'tips' dataset

```python
sns.set_theme(style='ticks', palette='pastel')
tips=sns.load_dataset('tips').dropna()
tips.head()
```

```python
sns.boxplot(
    data=tips,
    x='day',
    y='tip',
    hue='sex',
    palette=['m','g']
)
sns.despine(offset=10, trim=True)
```

## 3th example with 'planets' dataset

```python
planets=sns.load_dataset('planets')
planets.head()
```

```python
planets.info()
```

```python
fig, ax = plt.subplots(figsize=(7,6))
ax.set_xscale('log')

sns.boxplot(
    data=planets,
    x='distance',
    y='method',
    hue='method',
    whis=[0,100],
    width=6,
    palette='vlag'
)
ax.xaxis.grid(True)
ax.set(ylabel='')
sns.despine(trim=True, left=True)

sns.stripplot(
    data=planets,
    x='distance',
    y='method',
    size=4,
    color='.5'
)
```

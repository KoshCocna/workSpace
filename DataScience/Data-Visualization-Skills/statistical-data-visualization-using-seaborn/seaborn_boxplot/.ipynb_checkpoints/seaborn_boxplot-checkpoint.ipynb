```python
import seaborn as sns
from matplotlib import pyplot as plt
import warnings
```

```python
warnings.filterwarnings('ignore')
```

```python
mpg=sns.load_dataset('mpg').dropna()
mpg.head()
```

```python
mpg.nunique()
```

```python
sns.boxplot(data=mpg,x='model_year', y='mpg', hue='origin')
```

```python
sns.set_theme(style='ticks', palette='pastel')
```

```python
tips=sns.load_dataset('tips').dropna()
tips.head()
```

```python
sns.boxplot(
    data=tips,
    x='day',
    y='tip',
    hue='sex',
    palette=['m','g']
)
sns.despine(offset=10, trim=True)
```

```python
planets = sns.load_dataset('planets')
planets.head()
```

```python
planets.nunique()
```

```python
sns.set_theme(style='ticks', palette='pastel')
```

```python
fig, ax = plt.subplots(figsize=(7,6))
ax.set_xscale('log')
sns.boxplot(
    data=planets,
    x='distance',
    y='method',
    hue='method',
    whis=[0,100],
    width=6,
    palette='vlag'
)
sns.stripplot(
    planets,
    x='distance',
    y='method',
    size=4,
    color='.3'
)
ax.xaxis.grid(True)
ax.set(ylabel="")
sns.despine(trim=True, left=True)
```

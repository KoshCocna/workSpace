```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
import numpy as np
```

```python
warnings.filterwarnings('ignore')
```

```python
employee_salaries=pd.read_csv('Employee_Salaries.csv')
```

```python
employee_salaries.tail(3)
```

```python
employee_salaries.nunique()
```

```python
sns.displot(
    data=employee_salaries,
    x='Base_Salary',
    hue='Gender',
    col='Department',
    col_wrap=6,
    multiple='dodge',
    common_bins=False,
    facet_kws={'sharex': False, 'sharey': False}
)
```

```python
sns.set_theme(rc={'figure.figsize':(14,7)},style='white')
sns.histplot(
    data=employee_salaries,
    x='Base_Salary',
    hue='Gender',
    stat='density',
    multiple='dodge',
    kde=True
)
sns.despine()
```

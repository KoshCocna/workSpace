```python
def quality_level_by_size(x):
    if x < -4:
        return 'level-1'
    elif x < 0:
        return 'level-2'
    elif x < 4:
        return 'level-3'
    elif x < 8:
        return 'level-4'
    else:
        return 'level-none'
```

```python
banana['quality_level']=banana['Size'].apply(quality_level_by_size)
```

```python
banana.tail(3)
```

```python
sns.stripplot(
    data=banana, 
    x="Ripeness", 
    y="quality_level", 
    hue="Quality", 
    legend=False,
    dodge=True,
    jitter=False, 
    s=10, 
    marker="D", 
    linewidth=1, 
    alpha=.1,
)
```

```python
import pandas as pd
import seaborn as sns
import warnings
import matplotlib.pyplot as plt
```

```python
warnings.filterwarnings('ignore')
```

```python
raw_data=pd.read_csv('stocks.csv')
raw_data.head(3)
```

```python
raw_data.nunique()
```

```python
sns.lmplot(
    data=raw_data,
    x='Open',
    y='Close',
    hue='Ticker',
    height=4,
    aspect=1.5
)
```

```python
total_volume_by_ticker=raw_data.groupby(by=['Ticker'])['Volume'].sum()
total_volume_by_ticker=pd.DataFrame(total_volume_by_ticker, columns=['Volume']).reset_index()
total_volume_by_ticker
```

```python
fig,ax=plt.subplots(figsize=(4,6))
sns.barplot(
    data=total_volume_by_ticker,
    x='Ticker',
    y='Volume',
    ax=ax
)
plt.show()
```

```python
mean_volume_by_ticker=raw_data.groupby(by='Ticker')['Volume'].mean()
mean_volume_by_ticker=pd.DataFrame(mean_volume_by_ticker, columns=['Volume']).reset_index()
mean_volume_by_ticker
```

```python
fig,ax=plt.subplots(figsize=(4,4))
sns.barplot(
    data=mean_volume_by_ticker,
    x='Ticker',
    y='Volume',
    ax=ax
)
plt.show()
```

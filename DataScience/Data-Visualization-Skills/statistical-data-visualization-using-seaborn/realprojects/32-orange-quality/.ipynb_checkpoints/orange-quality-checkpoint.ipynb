```python
import pandas as pd
import seaborn as sns
import warnings
```

```python
warnings.filterwarnings('ignore')
```

```python
orange=pd.read_csv('Orange Quality Data.csv')
orange.head(3)
```

```python
orange.info()
```

```python
orange.nunique()
```

```python
orange.columns
```

```python
sns.lmplot(
    data=orange,
    x='Weight (g)',
    y='Brix (Sweetness)',
    hue='Color',
    col='Ripeness (1-5)',
    col_wrap=4,
    scatter_kws={'s':5},
    line_kws={'lw':1}
)
```

```python
g=sns.scatterplot(
    data=orange,
    x='HarvestTime (days)',
    y='Weight (g)',
    hue='Quality (1-5)'
)
sns.move_legend(g,'upper left', bbox_to_anchor=(1, 1), title='Quality')
sns.despine()
```

```python
sns.swarmplot(
    data=orange,
    x='Weight (g)',
    y='Color'
)
```

```python
sns.histplot(
    data=orange,
    x='Color',
    hue='Ripeness (1-5)'
)
```

```python
sns.set_theme(style="ticks")
palette = sns.color_palette("rocket_r")
sns.relplot(
    data=orange,
    x="Weight (g)", 
    y="Size (cm)",
    hue="Color", 
    #size="Softness (1-5)", 
    col="Ripeness (1-5)",
    col_wrap=4,
    kind="line", 
    palette=palette,
    height=5, 
    aspect=.75, 
    facet_kws=dict(sharex=False),
)
```

```python
sns.lmplot(
    data=orange,
    x='pH (Acidity)',
    y='Brix (Sweetness)'
)
```

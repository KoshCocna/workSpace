```python
import pandas as pd
import seaborn as sns
import warnings
```

```python
working_hours=pd.read_csv('annual-working-hours-per-worker.csv')
```

```python
# getting a new dataset to plot heatmap using pivot function 
```

```python
working_hours.head(3)
```

```python
working_hours=working_hours.pivot(index='Country', columns='Year', values='Average annual working hours per worker')
```

```python
working_hours
```

```python
sns.heatmap(
    data=working_hours
)
```

```python
filled_working_hours=working_hours.fillna(4000)
sns.heatmap(
    data=filled_working_hours,
    annot=True, 
    #fmt="d", 
    linewidths=.5
)
```

```python
raw_data=pd.read_csv('annual-working-hours-per-worker.csv')
raw_data
```

```python
new_df=pd.DataFrame(columns=['Country','Country Abbreviation','Year','Average annual working hours per worker'])
for i in ['Australia','Belgium','Canada','Denmark','France']:
    locals()[i]=raw_data[raw_data['Country']==i]
    new_df=pd.concat([new_df,locals()[i]],ignore_index=True)

new_df
sns.lineplot(
    data=new_df,
    x='Year',
    y='Average annual working hours per worker',
    hue='Country'
)
```

```python
new_df
new_df=new_df.rename(columns={'Average annual working hours per worker':'workhours'})
```

```python
new_df['Year']=new_df['Year'].astype('float')
new_df.info()
sns.lmplot(
    data=new_df,
    x='Year',
    y='workhours',
    hue='Country',
    aspect=2
)
```

```python
sns.scatterplot(
    data=new_df,
    x='Year',
    y='workhours',
    hue='Country'
)
```

```python
working_hours=working_hours.pivot(index='Country', columns='Year', values='1y_workinghours')
```

```python
working_hours
```

```python
sns.heatmap(
    data=working_hours
)
```

```python
filled_working_hours=working_hours.fillna(4000)
sns.heatmap(
    data=filled_working_hours,
    #annot=True, 
    #fmt="d", 
    linewidths=.5
)
```

```python
raw_data=pd.read_csv('annual-working-hours-per-worker.csv')
raw_data.rename(columns={'Average annual working hours per worker': '1y_work_hours'}, inplace=True)
raw_data.head(3)
```

```python
some_countries_df=pd.DataFrame(columns=['Country','Country Abbreviation','Year','1y_work_hours'])
for i in ['Australia','Belgium','Canada','Denmark','France']:
    locals()[i]=raw_data[raw_data['Country']==i]
    some_countries_df=pd.concat([some_countries_df,locals()[i]],ignore_index=True)

sns.lineplot(
    data=some_countries_df,
    x='Year',
    y='1y_work_hours',
    hue='Country'
)
```

```python
some_countries_df['Year']=some_countries_df['Year'].astype('float')
sns.lmplot(
    data=some_countries_df,
    x='Year',
    y='1y_work_hours',
    hue='Country',
    aspect=2
)
```

```python
sns.scatterplot(
    data=some_countries_df,
    x='Year',
    y='1y_work_hours',
    hue='Country'
)
```

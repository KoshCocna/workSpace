### cancer dataset

```python
cancer_b=pd.read_csv('cancer_b.csv')
```

```python
cancer_b
```

```python
cancer_b.info()
```

```python
cancer_b.nunique()
```

```python
cancer_mean=cancer_b[['Radius (mean)','Texture (mean)','Perimeter (mean)','Area (mean)','Smoothness (mean)','Compactness (mean)','Concavity (mean)','Concave points (mean)','Symmetry (mean)','Fractal dimension (mean)']]
```

```python
cancer_mean=cancer_mean.rename(columns={'Radius (mean)':'Radius','Texture (mean)':'Texture','Perimeter (mean)':'Perimeter','Area (mean)':'Area','Smoothness (mean)':'Smoothness','Compactness (mean)':'Compactness','Concavity (mean)':'Concavity','Concave points (mean)':'Concave points','Symmetry (mean)':'Symmetry','Fractal dimension (mean)':'Fractal dimension'})
```

```python
cancer_mean
```

```python
cancer_se=cancer_b[['Radius (se)','Texture (se)','Perimeter (se)','Area (se)','Smoothness (se)','Compactness (se)','Concavity (se)','Concave points (se)','Symmetry (se)','Fractal dimension (se)']]
```

```python
cancer_se=cancer_se.rename(columns={'Radius (se)':'Radius','Texture (se)':'Texture','Perimeter (se)':'Perimeter','Area (se)':'Area','Smoothness (se)':'Smoothness','Compactness (se)':'Compactness','Concavity (se)':'Concavity','Concave points (se)':'Concave points','Symmetry (se)':'Symmetry','Fractal dimension (se)':'Fractal dimension'})
```

```python
cancer_se
```

```python
cancer_worst=cancer_b[['Radius (worst)','Texture (worst)','Perimeter (worst)','Area (worst)','Smoothness (worst)','Compactness (worst)','Concavity (worst)','Concave points (worst)','Symmetry (worst)','Fractal dimension (worst)']]
```

```python
cancer_worst=cancer_worst.rename(columns={'Radius (worst)':'Radius','Texture (worst)':'Texture','Perimeter (worst)':'Perimeter','Area (worst)':'Area','Smoothness (worst)':'Smoothness','Compactness (worst)':'Compactness','Concavity (worst)':'Concavity','Concave points (worst)':'Concave points','Symmetry (worst)':'Symmetry','Fractal dimension (worst)':'Fractal dimension'})
```

```python
cancer_worst
```

```python
cancer_mean.insert(0,'status', 'mean')
```

```python
cancer_mean
```

```python
cancer_se.insert(0,'status','se')
cancer_worst.insert(0,'status','worst')
```

```python
cancer_info=pd.concat([cancer_mean, cancer_se, cancer_worst], ignore_index=True)
```

```python
cancer_info
```

```python
sns.scatterplot(
    data=cancer_info,
    x='Radius',
    y='Texture',
    hue='status',
    s=10
)
```

```python
sns.lmplot(
    data=cancer_info,
    x='Compactness',
    y='Area',
    hue='status',
    scatter_kws={"s": 10}, 
    line_kws={"lw":1}
)
```

```python
sns.kdeplot(
    data=cancer_info,
    x='Area'
)
```

```python
sns.lineplot(
    data=cancer_info,
    x=cancer_info.index,
    y='Symmetry'
)
```

```python
sns.stripplot(
    data=cancer_info,
    x='Compactness',
    y='Smoothness',
    hue='status',
    jitter=False,
    size=5,
    marker='D',
    linewidth=1,
    alpha=.3
)
```

```python
sns.catplot(
    data=cancer_info, 
    x="Concave points", 
    y="status", 
    kind="violin", 
    bw_adjust=.5, 
    cut=0, 
    split=True,
)
```

```python
sns.scatterplot(data=cancer_info, x="Radius", y="Area", hue='status', size=1)
sns.rugplot(data=cancer_info, x="Radius", y="Area", hue='status')
```

```python
sns.ecdfplot(data=cancer_info, x="Radius")
```

```python
sns.relplot(data=cancer_info, x="Concavity", y="Radius", hue="status")
```

```python
sns.set_theme(rc={'figure.figsize':(15,8)}, style='white')
sns.histplot(data=cancer_info, y="Area", binwidth=50, kde=True, multiple='dodge',hue='status')
```

```python
sns.set_theme(rc={'figure.figsize':(8,5)}, style='whitegrid')
sns.swarmplot(data=cancer_info, x="Perimeter", y="status", marker='x',linewidth=1)
```

```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
import warnings
import numpy as np
```

```python
warnings.filterwarnings('ignore')
```

```python
top_50_youtube_channels=pd.read_csv('Most Subscribed YouTube Channels_exported.csv')
```

```python
top_50_youtube_channels.head(3)
```

```python
top_50_youtube_channels.info()
```

```python
top_50_youtube_channels.nunique()
```

```python
plt.plot(np.arange(1,51).tolist(),top_50_youtube_channels['Subscribers (millions)'],'g^')
plt.axis((0,55,0,260))
plt.xlabel('channels')
plt.ylabel('subscribers')
plt.show()
```

```python
sns.scatterplot(
    data=top_50_youtube_channels,
    x=np.arange(1,51).tolist(),
    y='Subscribers (millions)',
    hue='Brand channel',
)
sns.despine()
```

```python
sns.displot(
    data=top_50_youtube_channels,
    x='Subscribers (millions)',
    hue='Brand channel',
    col='Category',
    col_wrap=4
)
```

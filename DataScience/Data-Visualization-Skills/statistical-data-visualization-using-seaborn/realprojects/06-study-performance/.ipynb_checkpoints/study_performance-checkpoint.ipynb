```python
import pandas as pd
import matplotlib.pyplot as plt 
import warnings
import seaborn as sns
```

```python
warnings.filterwarnings('ignore')
```

```python
raw_student_data=pd.read_csv('study_performance.csv')
```

```python
raw_student_data.head(3)
```

```python
raw_student_data.nunique()
```

```python
sns.lmplot(
    data=raw_student_data,
    x='math_score',
    y='reading_score',
    hue='gender',
    col='parental_level_of_education',
    col_wrap=3
)
```

```python
raw_student_data[raw_student_data['gender']=='female']['math_score'].sort_values()
```

```python
best_students=raw_student_data[(raw_student_data['math_score']==100) | (raw_student_data['reading_score']==100) | (raw_student_data['writing_score']==100)]
```

```python
best_students.head()
```

```python
best_students.info()
```

```python
best_students.nunique()
```

```python
sns.displot(
    data=best_students, 
    x="math_score", 
    hue="race_ethnicity", 
    multiple="stack", 
    #kind="kde",
    col='gender'
)
```

```python
best_students[best_students['race_ethnicity']=='group A']
```

```python
f, axs = plt.subplots(1, 2, figsize=(8, 4), gridspec_kw=dict(width_ratios=[4, 3]))
sns.scatterplot(data=raw_student_data, x="math_score", y="reading_score", hue="gender", ax=axs[0])
sns.histplot(data=raw_student_data, x="race_ethnicity", hue="race_ethnicity", shrink=.8, alpha=.8, legend=False, ax=axs[1])
f.tight_layout()
```

```python
g = sns.relplot(data=raw_student_data, x="reading_score", y="writing_score", col="gender")
g.set_axis_labels("Reading Score", "Writing Score")
```

```python
sns.jointplot(data=raw_student_data, x="math_score", y="reading_score", hue="parental_level_of_education")
```

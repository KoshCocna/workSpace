```python
import seaborn as sns
import warnings
import pandas as pd
import matplotlib.pyplot as plt
```

```python
warnings.filterwarnings('ignore')
```

```python
raw_Data=pd.read_csv('used_cars_data.csv',index_col='S.No.').dropna(how='all')
raw_Data.head(3)
```

```python
raw_Data.shape
```

```python
raw_Data.nunique()
```

```python
sns.set_style('white')
plt.rc('xtick', labelsize=14)
plt.rc('ytick', labelsize=14)
```

```python
sns.scatterplot(
    data=raw_Data,
    x='Year',
    y='Price',
    hue='Fuel_Type'
)
```

```python
new_data=raw_Data[['Price','Year']]
new_data.head(3)
```

```python
new_data['Year'].unique()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(new_data, 
             x='Year',
             #palette="viridis",
             kde=True)

sns.despine(left=True)
#plt.xlabel('')
#plt.yticks([])
plt.tight_layout();
```

```python
plt.figure(figsize=(8,5))
sns.histplot(new_data, 
             x='Year',
             y='Price',
             kde=True)
```

```python
sns.lmplot(
    data=raw_Data,
    y='Kilometers_Driven',
    x='Price',
    #hue='Owner_Type'
)
```

```python
raw_Data['Kilometers_Driven'].max()
```

```python
data_01=raw_Data[raw_Data['Kilometers_Driven']<300000]
data_01['Owner_Type'].unique()
```

```python
g = sns.relplot(
    x='Kilometers_Driven', 
    y='Price', 
    data=data_01,
    row='Owner_Type',
    row_order=['First', 'Second', 'Third', 'Fourth & Above'],
    hue='Fuel_Type',
    aspect=1.5,
    #s=300,
    style='Transmission',
    markers=['X', 'o']
)
g.set_titles(row_template="Owner_Type:{row_name}")
g.set_xlabels('Kilometers_Driven')
g.set_ylabels('Price')
leg = g._legend
leg.set_title('Car');
```

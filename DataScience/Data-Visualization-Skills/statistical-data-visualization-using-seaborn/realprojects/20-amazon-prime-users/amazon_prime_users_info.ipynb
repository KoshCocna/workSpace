```python
import pandas as pd
import seaborn as sns
import warnings
```

```python
warnings.filterwarnings('ignore')
```

```python
prime_users=pd.read_csv('amazon_prime_users.csv', index_col='User ID')
```

```python
prime_users.info()
```

```python
prime_users.nunique()
```

```python
prime_users.head(3)
```

```python
prime_users['Date of Birth']=pd.to_datetime(prime_users['Date of Birth'])
prime_users['Membership Start Date']=pd.to_datetime(prime_users['Membership Start Date'])
prime_users['Membership End Date']=pd.to_datetime(prime_users['Membership End Date'])
```

```python
prime_users.info()
```

```python
prime_users['Date of Birth'].dt.year
```

```python
sns.displot(
    data=prime_users,
    x=prime_users['Date of Birth'].dt.year,
    hue='Gender',
    multiple='dodge',
    col='Payment Information'    
)
```

```python
sns.scatterplot(
    data=prime_users,
    x=prime_users['Date of Birth'].dt.year,
    y='Feedback/Ratings'
)
```

```python
sns.swarmplot(
    data=prime_users,
    x=prime_users['Date of Birth'].dt.year,
    y='Favorite Genres'
)
```

```python
sns.lineplot(
    data=prime_users,
    x=prime_users['Date of Birth'].dt.year,
    y='Feedback/Ratings',
    hue='Devices Used'
)
```

```python
sns.histplot(
    data=prime_users,
    x='Favorite Genres',
    stat='density',
    kde=True
)
```

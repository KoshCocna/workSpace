```python
warnings.filterwarnings('ignore')
```

```python
raw_data=pd.read_csv('1000 companies list and rating.csv')
raw_data.head(3)
```

```python
raw_data.info()
```

```python
raw_data.nunique()
```

### convert age and reviewers columns to float data in order to analyse

```python
#rename a column name
df=raw_data.rename(columns={'Unnamed: 0':'index'})
#convert age column's type to float
df['age']=df['age'].str.replace(' years old', '').astype('float')
#convert reviewers column's type to float
df['reviewers']=df['reviewers'].str.replace(' Reviews','')
def value_to_float(x):
    if type(x) == float or type(x) == int:
        return x
    if 'k' in x:
        if len(x) > 1:
            return float(x.replace('k', '')) * 1000
        return 1000.0
    else:
        return float(x)
    return 0.0
df['reviewers'] = df['reviewers'].apply(value_to_float)
#set index as column index
df.set_index('index', inplace=True)
df.head(3)
```

### Now we are ready to analyse data

```python
top_100=df.head(100)
sns.lmplot(
    data=top_100,
    x='reviewers',
    y='rating'
)
```

```python
sns.lmplot(
    data=top_100,
    x='reviewers',
    y='age'
)
```

```python
sns.scatterplot(
    data=top_100,
    x=top_100.index,
    y='reviewers'
)
```

```python
sns.lineplot(
    data=top_100,
    x=top_100.index,
    y='rating',
)
```

```python
import pandas as pd
import seaborn as sns
import warnings
```

```python
warnings.filterwarnings('ignore')
```

```python
sleep_health=pd.read_csv('Sleep_health_and_lifestyle_dataset.csv',index_col='Person ID')
sleep_health.head(3)
```

```python
sleep_health.info()
```

```python
sleep_health.nunique()
```

```python
def get_high_pressure(pres):
    return pres.split('/')[0]
```

```python
sleep_health.insert(8,'High Blood Pressure', sleep_health['Blood Pressure'].apply(get_high_pressure).astype('int'))
```

```python
def get_low_pressure(pres):
    return pres.split('/')[1]
```

```python
sleep_health.insert(9, 'Low Blood Pressure', sleep_health['Blood Pressure'].apply(get_low_pressure).astype('int'))
```

```python
sleep_health.drop('Blood Pressure', axis=1, inplace=True)
```

```python
sleep_health.head()
```

```python
sleep_health.nunique()
```

```python
sns.scatterplot(
    data=sleep_health,
    x='Sleep Duration',
    y='Physical Activity Level',
    hue='Gender'
)
```

```python
sns.lineplot(
    data=sleep_health,
    x='Sleep Duration',
    y='Physical Activity Level',
    hue='Gender'
)
```

```python
sns.swarmplot(
    data=sleep_health,
    x='High Blood Pressure',
    y='BMI Category',
    hue='Gender'
)
```

```python
sns.histplot(
    data=sleep_health,
    x='Age',
    hue='Stress Level',
    multiple='dodge'
)
```

```python
sns.histplot(
    data=sleep_health,
    x='Daily Steps',
    kde=True,
    hue='BMI Category',
    multiple='dodge',
    palette={'Normal':'g','Normal Weight':'b','Overweight':'y','Obese':'r'}
)
```

```python
sns.lmplot(
    data=sleep_health,
    x='Sleep Duration',
    y='Heart Rate',
    hue='Gender'
)
```

```python
sns.lineplot(
    data=btc,
    x='Date',
    y='Close'
)
```

```python
btc_open=btc[['Date','Open']]
btc_open=btc_open.rename(columns={'Open':'Value'})
btc_open.insert(2, 'hue', 'open')
btc_close=btc[['Date', 'Close']]
btc_close=btc_close.rename(columns={'Close':'Value'})
btc_close.insert(2, 'hue', 'close')

btc_open_close=pd.concat([btc_open,btc_close], ignore_index=True)
btc_open_close
```

```python
sns.lineplot(
    data=btc_open_close,
    x='Date',
    y='Value',
    hue='hue',
    #style='hue',
    #markers=True,
    #markers='o'
)
```

```python
btc_2010=btc_open_close[btc_open_close['Date'].dt.year == 2010]
btc_2010.insert(3, 'year', 2010)
btc_2010
```

```python
btc_with_year_column=pd.DataFrame(columns=['Date','Value','hue','year'])
for i in range(2010, 2025):
    locals()['btc_'+str(i)]=btc_open_close[btc_open_close['Date'].dt.year == i]
    locals()['btc_'+str(i)].insert(3, 'year', i)
    btc_with_year_column=pd.concat([btc_with_year_column,locals()['btc_'+str(i)]], ignore_index=True)
btc_with_year_column
```

```python
sns.lineplot(
    data=btc_with_year_column,
    palette='tab10',
    linewidth=2.5,
    x='Date',
    y='Value',
    hue='year'
)
```

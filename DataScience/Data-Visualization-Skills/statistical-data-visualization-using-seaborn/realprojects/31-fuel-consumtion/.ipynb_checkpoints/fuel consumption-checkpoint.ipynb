```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
import warnings
```

```python
warnings.filterwarnings('ignore')
```

```python
cars=pd.read_csv('FuelConsumption (1).csv')
```

```python
cars.head(3)
```

```python
cars.info()
```

```python
cars.nunique()
```

```python
cars_2000=cars.drop('Year', axis=1)
cars_2000.tail()
```

```python
sns.histplot(
    data=cars_2000,
    x='CYLINDERS',
    hue='FUEL',
    multiple='dodge',
    kde=True,
    palette='Set2',
    element='step'
)
```

```python
sns.histplot(
    data=cars_2000,
    x='MAKE',
)
cars_2000['MAKE'].value_counts()
```

```python
sns.set_theme(style='white', rc={'figure.figsize':(12,5)})
sns.histplot(
    data=cars_2000,
    x='ENGINE SIZE',
    binwidth=0.2,
    hue='FUEL',
    palette={'X':'b','Z':'y','E':'g','N':'r','D':'purple'},
    multiple='dodge',
    kde=True
)
```

```python
cars_2000.columns
```

```python
cars_2000.rename(columns={'COEMISSIONS ':'COEMISSIONS'},inplace=True)
```

```python
sns.scatterplot(
    data=cars_2000,
    x='FUEL CONSUMPTION',
    y='COEMISSIONS',
    hue='CYLINDERS',
    size='FUEL'
)
```

```python
import pandas as pd
import seaborn as sns
import warnings
from matplotlib import pyplot as plt
```

```python
warnings.filterwarnings('ignore')
```

```python
raw_data=pd.read_csv('Stock_TLKM_2005-2024.csv')
raw_data.head(3)
```

```python
raw_data.insert(7,'year',raw_data['Date'].str[:4])
```

```python
raw_data
```

```python
data_by_year=raw_data.groupby('year')
```

```python
highest_by_Close_stock=pd.DataFrame(columns=raw_data.columns)
```

```python
for y, data in data_by_year:
    highest=data.nlargest(1, 'Close')
    highest_by_Close_stock=pd.concat(objs=[highest_by_Close_stock,pd.DataFrame(highest)])
highest_by_Close_stock.insert(8,'hue',0)
highest_by_Close_stock
```

```python
lowest_by_Close_stock=pd.DataFrame(columns=raw_data.columns)
```

```python
for x, data in data_by_year:
    lowest=data.nsmallest(1, 'Close')
    lowest_by_Close_stock=pd.concat(objs=[lowest_by_Close_stock,pd.DataFrame(lowest)])
lowest_by_Close_stock.insert(8,'hue',1)
lowest_by_Close_stock
```

```python
data1=pd.concat(objs=[highest_by_Close_stock,lowest_by_Close_stock])
data1.tail(3)
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
#sns.set(font_scale = 0.6)
g=sns.lineplot(
    data=data1,
    x='year',
    y='Close',
    hue='hue',
    ax=ax
)
plt.legend(title='max&min', fontsize=16)
#plt.xlabel('Year', fontsize=16);
plt.ylabel('Stock Price', fontsize=16);
plt.title('Tele-Company Stock Price', fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=8)
```

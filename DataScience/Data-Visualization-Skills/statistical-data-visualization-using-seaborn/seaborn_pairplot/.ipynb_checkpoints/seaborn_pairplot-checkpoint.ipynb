```python
import seaborn as sns
import warnings
```

```python
warnings.filterwarnings('ignore')
sns.set_theme(style='ticks')
```

```python
penguins=sns.load_dataset('penguins').dropna()
penguins.head(3)
```

```python
sns.pairplot(
    data=penguins,
    hue='species'
)
```

```python
iris=sns.load_dataset('iris').dropna()
iris.tail(3)
```

```python
sns.pairplot(
    data=iris,
    hue='species'
)
```

```python
import plotly.express as px
import pandas as pd
```

```python
data=pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/gapminder_with_codes.csv')
```

```python
data.head(3)
```

```python
data.info()
```

```python
fig=px.choropleth(
    data,
    locations='iso_alpha',
    color='gdpPercap',
    hover_name='country',
    projection='natural earth',
    animation_frame='year',
    title='GDP per Capital by Country'
)
fig.show()
```
